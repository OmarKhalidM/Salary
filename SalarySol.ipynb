## Transform The Opject Data TO Numeric
***

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for col in da.columns:
    
    if da[col].dtype=='object':
        da[col]=le.fit_transform(da[col])
        da[col]=pd.to_numeric(da[col])
```

```python
da.info()
```

## See The Destripution of the Data
***

```python
i=1
fig,axs=plt.subplots(1,3,figsize=(20,20))   
for col in da.columns:
    plt.subplot(4,4,i)
    sb.distplot(da[col],kde=False)
    i+=1
plt.tight_layout(pad=0.8)
plt.show()
```

```python
i=1
fig,axs=plt.subplots(1,3,figsize=(20,20))   
for col in da.columns:
    plt.subplot(4,4,i)
    plt.scatter(x=da[col],y='salary',data=da)
    plt.xlabel(col)
    i+=1
plt.tight_layout(pad=0.8)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sb.heatmap(da.corr(),annot=True)
plt.show()
```

# Applay Traning on data
***

```python
x=da.iloc[:,:-1].values #get All col without the last one
y=da.iloc[:,-1].values
```

***
### Frist split the data to train and test  by (80%) to train and (20%) to test
***

```python
from sklearn.model_selection import train_test_split

train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
```

***
## Apply The linear Regriton model
***

```python
from sklearn.linear_model import LinearRegression
model =LinearRegression()
model.fit(train_x,train_y)
```

```python
y_pred=model.predict(test_x)
```

***
## Teat the error Rate
***

```python
from sklearn.metrics import mean_squared_error ,r2_score
print ('The Mean Squared Error :',mean_squared_error(test_y,y_pred))
```

```python
print("THe R2Score : ",r2_score(test_y,y_pred))
```
